```python
import pandas as pd
import numpy as np
import pandas_profiling
import torch 
from torch.autograd import Variable 
from torch import nn, optim
import random 
from operator import add
import torch
from torch.utils.data import DataLoader
```

```python
NUM_OF_VAR=100
```

```python
x = random.sample(range(-500, 500), NUM_OF_VAR)
y = list(map(lambda y : y + 50, x)) 
```

```python
print(x[0:10])
print(y[0:10])
```

```python
x=Variable(torch.Tensor(x)).view(NUM_OF_VAR,1)
y=Variable(torch.Tensor(y)).view(NUM_OF_VAR,1)
```

```python
class RegressionDataset(torch.utils.data.Dataset):
    def __init__(self, x, y):
        super(RegressionDataset, self).__init__()
        self._x = x
        self._y = y

    def __getitem__(self, i):
        return self._x[i],self._y[i]

    def __len__(self):
        return len(self._x)

    def __iter__(self):
        for x in self._x:
            yield x

    def get_labels(self):
        return self._y

```

```python
reg_dataset = RegressionDataset(x,y)
```

```python
reg_dataset[0]
```

```python
BATCH_SIZE=16
```

```python
data = DataLoader(reg_dataset, batch_size=BATCH_SIZE, shuffle=True)
```

```python



class LinearRegressionModel(torch.nn.Module): 
  
    def __init__(self, input_size, hidden_size ,output_size): 
        super(LinearRegressionModel, self).__init__() 
        self.linear1 = torch.nn.Linear(input_size, hidden_size)  # One in and one out 
        self.relu1= torch.nn.ReLU()
        #self.Norm1 = nn.BatchNorm1d(1000,1000)
        self.linear2 = torch.nn.Linear(hidden_size,hidden_size)
        self.relu2= torch.nn.ReLU()
        
        self.linear3 = torch.nn.Linear(hidden_size,output_size)
  
    def init_weights(self):
        initrange = 0.1
        self.linear1.weight.data.uniform_(-initrange, initrange)        
        #self.linear1.bias.data.zero_()
        self.linear2.weight.data.uniform_(-initrange, initrange)        
        #self.linear2.bias.data.zero_()
        self.linear3.weight.data.uniform_(-initrange, initrange) 
        #self.linear4.weight.data.uniform_(-initrange, initrange) 
        #self.linear3.bias.data.zero_()
     
    def forward(self, x): 
        lin1 = self.linear1(x)
        #norm1=self.Norm1(lin1)
        reul=self.relu1(lin1)
        lin2 = self.linear2(reul)
        reul2=self.relu1(lin2)
        lin3 = self.linear3(reul2)
        y_pred=lin3
        return y_pred 
```

```python
class Linear_regression(nn.Module):
    
    # Constructor
    def __init__(self, input_size, hidden_size ,output_size):
        super(Linear_regression, self).__init__()
        self.linear1 = nn.Linear(input_size, hidden_size,bias=True)
        self.relu1= torch.nn.ReLU()
        self.linear2 = nn.Linear(hidden_size, hidden_size,bias=True)
        self.relu2= torch.nn.ReLU()
        self.linear3 = nn.Linear(hidden_size, output_size,bias=True)
        

        
    # Prediction
    def forward(self, x):
        lin1=self.linear1(x )
        #torch.nn.init.xavier_normal_(lin1, gain=1.0)
        relu1= self.relu1(lin1)
        lin2=self.linear2(relu1)
        #torch.nn.init.xavier_normal_(lin2, gain=1.0)
        relu2= self.relu2(lin2)
        yhat = self.linear3(relu2)
        return yhat
```

```python
input_size=1
output_size=1
hidden_size=32
model = LinearRegressionModel(input_size, hidden_size,output_size)
N_EPOCHS=2000
#model = LinearRegressionModel()
```

```python

#criterion = torch.nn.MSELoss(size_average = False) 
criterion = torch.nn.MSELoss() #L1Loss()## MSELoss() 

optimizer = torch.optim.Adam(model.parameters(),lr = 0.001) ##   SGD(model.parameters(), lr = 0.01) 
```

```python
#data,for text, offsets, cls in data
for epoch in range(N_EPOCHS):
    for x,y in data:
        pred_y = model(x) 
        loss = criterion(pred_y, y) 
        optimizer.zero_grad() 
        loss.backward() 
        optimizer.step() 
    print('epoch {}, loss {}'.format(epoch, loss.item())) 
```
